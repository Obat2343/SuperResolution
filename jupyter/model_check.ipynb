```python
config_file_path = '../configs/test.yaml'
cfg.merge_from_file(config_file_path)
cfg.freeze()
torch.manual_seed(cfg.SEED)
cuda = torch.cuda.is_available()
if cuda:
    torch.backends.cudnn.benchmark = True
    torch.backends.cudnn.deterministic = True
    torch.cuda.manual_seed(cfg.SEED)

print('Running with config:\n{}'.format(cfg))
```

```python
device = torch.device(cfg.MODEL.DEVICE)
gen_pretrain_path = '../output/test_128_256_512_clip_2/generator/iteration_30000.pth'
dis_pretrain_path = '../output/test_128_256_512_clip_2/discriminator/iteration_30000.pth'

print('Building model...')
model = build_model(cfg).to(device)

print('--------------------Generator architecture--------------------')
print(model.sr_model)
print('------------------Discriminator architecture------------------')
print(model.discriminators)

print('Loading datasets...')
train_transform = TrainAugmentation(cfg)
train_dataset = build_dataset(dataset_list=cfg.DATASETS.TRAIN, transform=train_transform)

eval_transform =  EvaluateAugmentation(cfg)
eval_dataset = build_dataset(dataset_list=cfg.DATASETS.EVAL, transform=eval_transform)

minitrain_transform = MiniTrainAugmentation(cfg)
minitrain_dataset = build_dataset(dataset_list=cfg.DATASETS.VIS, transform=minitrain_transform)

# if args.sync_batchnorm:
#     model = convert_model(model).to(device)

# if args.mixed_precision:
#     model.sr_model, optimizer = apex.amp.initialize(model.sr_model, optimizer, opt_level='O1')
#     if not args.pretrain:
#         model.discriminators, d_optimizer = apex.amp.initialize(model.discriminators, d_optimizer, opt_level='O1')

model.sr_model.load_state_dict(fix_model_state_dict(torch.load(gen_pretrain_path, map_location=lambda storage, loc:storage)))
model.discriminators.load_state_dict(fix_model_state_dict(torch.load(dis_pretrain_path, map_location=lambda storage, loc:storage)))
```

```python
from IPython.display import display
import torchvision
discriminator_name_list = list(cfg.SOLVER.DISCRIMINATOR_TYPE)
discriminator_name_list.sort()
    
topil = torchvision.transforms.ToPILImage(mode=None)

model.eval()
lr_images, hr_images, filename = train_dataset[2]
lr_images, hr_images = torch.unsqueeze(lr_images,0).to(device), torch.unsqueeze(hr_images,0).to(device)

with torch.no_grad():
    sr_images, sr_prediction, hr_prediction = model.debug_forward(lr_images, hr_images)
    
lr_image = topil(lr_images[0].cpu())
display(lr_image)

hr_image = topil(hr_images[0].cpu())
display(hr_image)
for i,disc_name in enumerate(discriminator_name_list):
    print('disc:{} prob:{}'.format(disc_name,hr_prediction[i].item()))

sr_image = topil(sr_images[0].cpu())
display(sr_image)
for i,disc_name in enumerate(discriminator_name_list):
    print('disc:{} prob:{}'.format(disc_name,sr_prediction[i].item()))
```
